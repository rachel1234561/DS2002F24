```python
import pandas as pd
import json
import requests
```

```python
#This function takes the input of the abreviation for a stock and returns Ticker Name, Full Name of the Stock,
#Current Market Price, Target Mean Price, 52 week high and 52 week low or reports that there is an error if needed

def get_stock_info(stock, apikey=''):
  url = "https://yfapi.net/v6/finance/quote"
  querystring = {"symbols": stock}
  headers = {'x-api-key': apikey}
  response = requests.request("GET", url, headers=headers, params=querystring)

  url2 = "https://yfapi.net/v11/finance/quoteSummary/" + stock + "?modules=financialData"
  response2 = requests.request("GET", url2, headers=headers, params=querystring)

  stock_json = response.json()
  stock2_json = response2.json()

#I used the build in AI feature on colab to figure out the syntax making a try-except sequence

  try:
    print("Ticker:", stock)
    stock_info = stock
    full_name = stock_json['quoteResponse']['result'][0]["longName"]
    print("Full Name:", full_name)
    current_price = stock_json['quoteResponse']['result'][0]["regularMarketPrice"]
    print("Current Market Price: $", current_price)
    stock_info += " " + str(current_price)

    if 'quoteSummary' in stock2_json and 'result' in stock2_json['quoteSummary'] and stock2_json['quoteSummary']['result']:
       target_mean_price = stock2_json['quoteSummary']['result'][0]['financialData']['targetMeanPrice']['raw']
       print("Target Mean Price:", target_mean_price)
       stock_info += " " + str(target_mean_price)
    else:
       print("Target Mean Price: Not Found")

    week52_high = stock_json['quoteResponse']['result'][0]['fiftyTwoWeekHigh']
    print("52 Week High: $", week52_high)
    stock_info += " " + str(week52_high)
    week52_low = stock_json['quoteResponse']['result'][0]['fiftyTwoWeekLow']
    print("52 Week Low: $", week52_low)
    stock_info += " " + str(week52_low)
    return stock_info
  except KeyError:
    print("Error: Could not retrieve stock information. Please check the stock try again.")
```

```python
#ask the user to enter a stock and use the get_stock_info to display the information on the stock
user_stock=input()
print(get_stock_info(user_stock))

url = "https://yfapi.net/v1/finance/trending/US"
headers = {'x-api-key': ''}
response = requests.request("GET", url, headers=headers)

data_trend = response.json()

print("Trending stocks:")
print(data_trend['finance']['result'][0]['quotes'][3]['symbol'])
print(data_trend['finance']['result'][0]['quotes'][6]['symbol'])
print(data_trend['finance']['result'][0]['quotes'][7]['symbol'])
print(data_trend['finance']['result'][0]['quotes'][14]['symbol'])
print(data_trend['finance']['result'][0]['quotes'][15]['symbol'])
```

```python
url = "https://yfapi.net/v1/finance/trending/US"
headers = {'x-api-key': ''}
response = requests.request("GET", url, headers=headers)

data_trend = response.json()

#adds info on trending stocks to the list trending_stocks
trending_stocks = []
trending_stocks.append(data_trend['finance']['result'][0]['quotes'][3]['symbol'])
trending_stocks.append(data_trend['finance']['result'][0]['quotes'][6]['symbol'])
trending_stocks.append(data_trend['finance']['result'][0]['quotes'][7]['symbol'])
trending_stocks.append(data_trend['finance']['result'][0]['quotes'][14]['symbol'])
trending_stocks.append(data_trend['finance']['result'][0]['quotes'][15]['symbol'])
```

```python
#makes a list that inclused the information on both the trending stocks and the stock the user input
stock_infos = []
for stock in trending_stocks:
  stock_info = get_stock_info(stock)
  if stock_info:
    stock_infos.append(stock_info)

stock_data = get_stock_info(user_stock)
stock_infos.append(stock_data)
```

```python
#makes the list of stock information into a pandas df
df = pd.DataFrame([x.split() for x in stock_infos])
```

```python
#I used the build in ai feature to add colunm names to my df
df.columns = ['Ticker', 'Current Price', 'Target Mean Price', '52 Week High', '52 Week Low']
df
```

```python
#makes the df into a csv and saves it
df.to_csv('stock_data.csv', index=False)

print("CSV file 'stock_data.csv' created successfully!")
```
